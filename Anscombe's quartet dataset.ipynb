## Background 

The Anscombe quartet dataset comprised of four distinct datasets each with nearly identical statistical characteristics 
such as mean, correlation and regression lines. 

In 1973 Francis J. Anscombe puplished a paper named Graphs in Statistical Analysis that contained these four datasets. The paper
shows that even though the summary statistics are almost identical, but when plotted the graphs show that the datasets
are actually quite different, giving the user a very perspective on the dataset. 

The paper demonstrated the importance of graphs in analysisng data in conjunction with numeric data. 

"A computer should make both calculations and graphs. Both sorts of output should be studied; 
each will contribute to understanding."

Francis J. Anscombe was an English statistician educated at Cambridge University.

### Discussion of plots and statistics above

From the numeric descriptive statistics above one would think that each datset is almost identical, if not duplicted. If one was to summarise their analysis on based on these statistics alone it could to a misleading conclusion. For example the coefficient of 0.82 indicates a strong relationship between each of the variables for each dataset, along with a similar standard deviation would lead one to believe that there is a linear relationship between the variables in each dataset. 

However, once the points for each dataset are plotted the conclusion drawn from the analysis becomes very different. 

Starting with X1 Y1 dataset the the realtionship here is possibly linear as each point is following the line.

The X2 Y2 dataset is not linear as the datapoints take on a curved shape.

The X3 Y3 dataset at initial inspection look linear but then an outlier is visible (12.74).

And in the final dataset X4 Y4 the relationship between the variables is vertical straight line with an outlier.
