# Investigation of the Anscombe's quartet dataset 

## Background 

The Anscombe quartet dataset comprised of four distinct datasets each with nearly identical statistical characteristics 
such as mean, correlation and regression lines. 

In 1973 Francis J. Anscombe puplished a paper named Graphs in Statistical Analysis that contained these four datasets. The paper
shows that even though the summary statistics are almost identical, but when plotted the graphs show that the datasets
are actually quite different, giving the user a very perspective on the dataset. 

The paper demonstrated the importance of graphs in analysisng data in conjunction with numeric data. 

"A computer should make both calculations and graphs. Both sorts of output should be studied; 
each will contribute to understanding."

Francis J. Anscombe was an English statistician educated at Cambridge University.

```python
#Import seaborn package for visualizations
import seaborn as sns
```

```python
# Import variables, dataset, change colours of scatterline and plots
V1 = sns.regplot(x="X1", y="Y1", data=df2,ci=None, color = 'blue',scatter_kws={"s": 80},line_kws={"color":"r","lw":2})
V1.axes.set_title('X1_Y1')
```

```python
V2 = sns.regplot(x="X2", y="Y2", data=df2,ci=None, color = 'blue',scatter_kws={"s": 80},line_kws={"color":"r","lw":2})
V2.axes.set_title('X2_Y2')
```

```python
V3 = sns.regplot(x="X3", y="Y3", data=df2,ci=None, color = 'blue',scatter_kws={"s": 80},line_kws={"color":"r","lw":2})
V3.axes.set_title('X3_Y3')
```

```python
V3 = sns.regplot(x="X4", y="Y4", data=df2,ci=None, color = 'blue',scatter_kws={"s": 80},line_kws={"color":"r","lw":2})
V3.axes.set_title('X4_Y4')
```

```python
# import matlotlip package to create a grid with containing the plots above. change marker size using scatter_kws 
import matplotlib.pyplot as plt
plt.subplot(2, 2,1)
V1 = sns.regplot(x="X1", y="Y1", data=df2,ci=None, color = 'blue',scatter_kws={"s": 20},line_kws={"color":"r","lw":2})
V1.axes.set_title('X1_Y1')
plt.subplot(2, 2,2)
V2 = sns.regplot(x="X2", y="Y2", data=df2,ci=None, color = 'blue',scatter_kws={"s": 20},line_kws={"color":"r","lw":2})
V2.axes.set_title('X2_Y2')
plt.subplot(2, 2,3)
V3 = sns.regplot(x="X3", y="Y3", data=df2,ci=None, color = 'blue',scatter_kws={"s": 20},line_kws={"color":"r","lw":2})
V3.axes.set_title('X3_Y3')
plt.subplot(2, 2,4)
V3 = sns.regplot(x="X4", y="Y4", data=df2,ci=None, color = 'blue',scatter_kws={"s": 20},line_kws={"color":"r","lw":2})
V3.axes.set_title('X4_Y4')
#subplots adjust changes hight and width of the space between the plots
plt.subplots_adjust(top=0.95, bottom=0.01, left=0.10, right=0.95, hspace=0.6,
                    wspace=0.35,)
```
